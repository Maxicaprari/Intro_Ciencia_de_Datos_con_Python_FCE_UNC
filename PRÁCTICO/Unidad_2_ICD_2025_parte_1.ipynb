1.   **Tipos de Datos**

üìå 1. Tipos de datos: estructurados y no estructurados
Datos estructurados: organizados en tablas con filas y columnas. Ej: bases de datos relacionales, archivos CSV, Excel.

Datos no estructurados: sin formato predefinido. Ej: texto libre, im√°genes, audios, videos, publicaciones en redes sociales.

**Importancia**: Identificar el tipo de dato es clave para elegir herramientas y t√©cnicas de procesamiento. En ciencia de datos, muchas veces se combinan ambos.


-----------
Antes de realizar cualquier an√°lisis, se debe garantizar que los datos sean confiables.
Es una etapa cr√≠tica del **Data Wrangling** dentro del ciclo de Ciencia de Datos. Es el proceso de tomar datos crudos y transformarlos en un formato limpio, estructurado y utilizable.
Es como ‚Äúdomesticar los datos‚Äù (de ah√≠ la palabra "wrangling", que viene del mundo del rodeo). En la pr√°ctica, los datos nunca vienen perfectos: pueden estar incompletos, desordenados, en el formato incorrecto, o tener errores. Data Wrangling es el conjunto de tareas y t√©cnicas que permiten limpiar, transformar y preparar esos datos.

üß∞ Incluye tareas como:
*   Limpieza de datos

*   Eliminar duplicados

*   Corregir errores tipogr√°ficos o de formato

*   Rellenar, eliminar o imputar valores faltantes

*   Conversi√≥n de formatos

*   Pasar fechas de string a datetime

*   Convertir columnas categ√≥ricas a num√©ricas (encoding)

*   Filtrado y selecci√≥n

*   Elegir solo las columnas o filas relevantes

*   Aplicar condiciones para depurar los datos

*   Fusi√≥n e integraci√≥n de fuentes

*   Combinar m√∫ltiples datasets

*   Unir informaci√≥n complementaria

*   Transformaciones

*   Crear nuevas variables

*   Escalar o normalizar datos

*   Agrupar o pivotear tablas

*   Otras...

üßπ 2. Limpieza e integraci√≥n de datos
**Limpieza**: detecci√≥n y correcci√≥n de errores (valores faltantes, duplicados, tipos err√≥neos).

**Integraci√≥n**: unificaci√≥n de datos de diferentes fuentes, garantizando coherencia y calidad.


üß¨ 3. T√©cnicas de curaci√≥n de datos e ingenier√≠a de caracter√≠sticas (feature engineering)
Curaci√≥n de datos: proceso continuo de mantenimiento y mejora de la calidad de los datos.

Feature engineering: creaci√≥n, transformaci√≥n o selecci√≥n de variables relevantes para modelos anal√≠ticos o predictivos.

üëâ Clave: Estas pr√°cticas aumentan el valor de los datos y mejoran la performance de modelos de Machine Learning.

üìè 4. Estandarizaci√≥n de procesos
Implica definir y aplicar reglas consistentes para el tratamiento de datos (nombres de columnas, formatos, unidades, etc.).

Mejora la reproducibilidad, escalabilidad y colaboraci√≥n entre equipos.

üëâ Contexto: En el ciclo de Ciencia de Datos, permite replicar experimentos, construir pipelines robustos y facilitar auditor√≠as.

5. Principios en el manejo de bases de datos
Incluye: normalizaci√≥n, relaciones entre tablas, consultas eficientes, integridad de datos.

Orienta sobre c√≥mo dise√±ar, consultar y mantener una base de datos relacional o no relacional.

üëâ Relevancia: Una buena base de datos facilita todo el flujo posterior de an√°lisis, visualizaci√≥n y modelado.

üß™ 6. Uso de Python y librer√≠as principales
pandas: manipulaci√≥n de datos tabulares.

*numpy*: operaciones matem√°ticas eficientes, vectores y matrices.

*scipy*: herramientas cient√≠ficas y estad√≠sticas.

Estas librer√≠as permiten implementar todo lo anterior: limpieza, transformaci√≥n, integraci√≥n y an√°lisis, de forma program√°tica y reproducible.

‚úÖ M√©todo: Es una funci√≥n que est√° dentro de un objeto o clase, y se llama como *objeto.metodo()*

Funciones comunes de io:
io.BytesIO():
Crea un buffer en memoria que act√∫a como si fuera un archivo, pero los datos se almacenan en la memoria en lugar de en el disco. Se usa mucho cuando se manipulan archivos binarios en memoria.

io.StringIO():
Similar a BytesIO(), pero en lugar de trabajar con datos binarios, se usa para trabajar con cadenas de texto en memoria. Es √∫til cuando se necesita trabajar con texto como si fuera un archivo.

ectura y escritura en archivos:
El m√≥dulo io tambi√©n puede ser √∫til cuando se trabajan con archivos en modo binario o texto, proporcionando m√°s control sobre la codificaci√≥n y los flujos de datos.

```python
url = 'http://cdn.buenosaires.gob.ar/datosabiertos/datasets/ministerio-de-economia-y-finanzas/sueldo-funcionarios/Sueldo_Funcionarios_2023.csv'

# Lo almacenamos en una variable llamada salarios
salarios = pd.read_csv( url,
                        encoding='ISO-8859-1', # Esta codificaci√≥n evita errores con caracteres especiales
                        on_bad_lines="skip", # Omitir l√≠neas problem√°ticas que podr√≠an interrumpir la carga del archivo
                        sep=";" # Usar el punto y coma (;) como delimitador
                        )
```

```python
salarios.head()
```

El par√°metro url es el que indica el nombre (en este caso la direcci√≥n web) del archivo a leer. Los otros par√°metros son opcionales y tiene que ver m√°s con el tipo y formato que tienen los datos de este archivo en particular. Pueden ver m√°s informaci√≥n sobre la funci√≥n read_csv en la documentaci√≥n oficial de pandas read_csv.
