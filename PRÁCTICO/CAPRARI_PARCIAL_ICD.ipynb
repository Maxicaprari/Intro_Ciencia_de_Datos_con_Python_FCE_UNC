# 3 ANALISIS

```python
#DETECTAR VALOSRES NULOS

df.isnull().sum()


```

```python
#AHORA EN PORCENTAJES
datanull = df.isnull().sum()
data_porcentaje = (datanull / len(df)) * 100
print("Porcentaje de datos faltantes por columna:\n", data_porcentaje)
```

```python
umbral = 0.30
dflimpio = df.dropna(thresh=int((1-umbral)*df.shape[0]), axis=1)
dflimpio.head()
```

```python
dflimpio.columns
```

Eliminaria la columna ISO3166_21 debido a que supera el umbral de datos faltantes preestablecido en 0.3

```python
#CAMBIO EL TIPO DE DATO DE FACTURACION
df['facturacion'] = df['facturacion'].astype(float)
df.dtypes
```

```python
df["facturacion_promedio_por_operacion"] = df['facturacion']/df["operaciones"]
df.head()
```

```python
df["densidad_poblacional"] = df['poblacion_proy_2020']/df['sup_km2']
df.head()
```

# ANALISIS EXPLORATORIO SENCILLO

```python
facturacion_promedio_total = df['facturacion'].mean()
print("Facturación promedio total: ", facturacion_promedio_total)
```

```python
facturacion_promedio_por_provincia = df.groupby('provincia')['facturacion'].mean()

facturacion_promedio_por_provincia
```

```python

min_facturacion = df['facturacion'].min()
max_facturacion = df['facturacion'].max()
rango = max_facturacion - min_facturacion
print("Rango de facturación: ", rango)
```
