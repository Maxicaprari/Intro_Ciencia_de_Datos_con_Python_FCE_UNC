## Introducción a la programación

### ¿Qué es un lenguaje de programación?

<p style='text-align: justify;'> Una computadora digital comprende y trabaja con valores binarios, es decir, unos y ceros. Sin embargo, para interactuar con los humanos y permitirles escribir instrucciones, se utilizan lenguajes de programación que son más legibles y comprensibles para nosotros. Un lenguaje de programación es un conjunto de reglas y sintaxis que permite a los programadores escribir instrucciones y comandos que una computadora puede entender y ejecutar. Es un medio de comunicación entre los humanos y las máquinas, permitiendo a los programadores expresar sus ideas y soluciones en un formato comprensible por las computadoras.

### ¿Qué es el código fuente?

El código fuente es ese texto legible por humanos que contiene las instrucciones escritas en un lenguaje de programación específico. Los programadores escriben el código fuente utilizando palabras clave, estructuras y sintaxis determinadas. Luego, se utiliza un programa especial llamado compilador o intérprete para traducir ese código fuente a un lenguaje de bajo nivel o código máquina, compuesto por instrucciones en forma de unos y ceros, que la computadora puede entender y ejecutar.

En resumen, el código fuente actúa como un puente entre los programadores y las computadoras, ya que permite expresar las instrucciones en un formato que los humanos pueden entender y luego ser traducido al lenguaje de la máquina para que la computadora pueda ejecutarlas.

### ¿Qué es programar?

Programar es el proceso de dar instrucciones a una computadora para que realice tareas específicas. Es como escribir un conjunto de pasos o reglas en un lenguaje especial que la computadora puede entender y seguir. Los programadores utilizan lenguajes de programación para comunicarse con la computadora y decirle qué hacer. Al programar, se pueden crear programas, aplicaciones y software que nos ayudan a resolver problemas, automatizar tareas y realizar diversas funciones.

### Operadores de **pertencia**
Los operadores *in* y *not in* se utilizan para comprobar si un elemento pertenece o no a una estructura de datos como listas, tuplas, diccionarios o cadenas de texto.

Operador *in* (pertenencia)
* ✅ Devuelve True si el elemento está presente en la estructura.
* ✅ Devuelve False si el elemento NO está presente.

Operador *not in* (no pertenencia)
* ✅ Devuelve True si el elemento NO está presente.
* ✅ Devuelve False si el elemento SÍ está presente.

```python
# Operadores de pertenencia
list1= [1,2,3,4,5] ## Lista de números
string1= 'Esto es una string.' ## String
tuple1= ("Esto", "es", "una", "tupla.") # Tupla de palabras
## Uso del operador "in"
print('Operador In:')
print(4 in list1)
print("es" in string1)
print("es" in tuple1)

# Separador visual en la salida
print(30*'-')

# Uso del operador "not in"
print('Operador Not In:')
print(1 not in list1)
print('Hola' not in list1)
print('Hola' not in tuple1)
```

Conclusión

**in** → Comprueba si un elemento está presente en una estructura de datos.

**not in** → Comprueba si un elemento NO está presente en una estructura de datos.

### Estructuras de Control

Las estructuras de control en Python son bloques de código que permiten tomar decisiones, repetir acciones y ejecutar diferentes secciones de código según ciertas condiciones.

#### Estructuras de Control Condicionales

**Sentencias 'if-elif-else'**

Permite evaluar múltiples condiciones en orden y ejecutar diferentes bloques de código según cada condición.

```python
a = 200
b = 300

if b > a:
    print("B es mayor que A")  # Se ejecuta esta línea porque 300 > 200
elif a == b:
    print("A y B son iguales")  # No se ejecuta porque 200 ≠ 300
else:
    print("A is mayor que B")  # No se ejecuta porque 200 < 300
```

#### Estructuras de Bucle

**Blucle 'while'**

Permite repetir un bloque de código mientras se cumpla una condición específica.

```python
j = 1
i = 1
while i <= 5:
    print(i)  # Imprime 1 infinitamente porque i no cambia
    j += 1  # Se incrementa j, pero no i (esto no afecta la condición)
else:
    print('Stop!')
```

El bloque anterior tiene un error lógico porque la variable i no se está incrementando dentro del bucle while. Provocando un bucle infinito ya que **i** siempre será 1 y nunca llegará a 5

```python
i = 1
while i <= 5:
    print(i)
    i += 1 # i = i + 1
print("Stop!")
```

```python
j = 1
i = 1
while i <= 5:
    print(i)  # Imprime 1 infinitamente porque i no cambia
    j += 1  # Se incrementa j, pero no i (esto no afecta la condición)
else:
    print('Stop!')
```

**Bucle 'for'**

Permite iterar sobre una secuencia de elementos (como una lista, una cadena de caracteres, etc.) y ejecutar un bloque de código para cada elemento.

```python
fruits = ["Manzana", "Banana", "Frutilla"]
for fruta in fruits:   #Se usa un for para recorrer la lista. En cada iteración, la variable fruta tomará un valor de la lista
  if fruta == "Banana":
    print("Ecuador")
  print(fruta)
```

Veamos en el bloque anterior como el if verifica si la fruta actual es "Banana". Si lo es, imprime "Ecuador".
Luego, siempre imprime la fruta actual, sin importar si entró o no al if.

# Ciencia de Datos

## 1. Definición y rol en la actualidad
La Ciencia de Datos es un campo interdisciplinario que integra estadística, matemáticas, informática y conocimiento del dominio para extraer información útil a partir de datos. Su meta principal es convertir grandes volúmenes de datos en conocimiento accionable mediante análisis, visualización y modelado.

**Importancia en la actualidad**  
Vivimos en una era orientada a los datos. La toma de decisiones basada en datos (`data-driven decision-making`) es central en negocios y políticas públicas.  
Herramientas como machine learning, analítica avanzada y dashboards permiten decisiones rápidas y fundamentadas.  
La gestión eficiente de datos (limpios, actualizados, estructurados) es clave para obtener buenos resultados.

## Aplicaciones en Ciencias Económicas
- **Análisis de mercado**: segmentación de clientes, identificación de patrones de consumo.
- **Predicción de tendencias**: precios, inflación, demanda.
- **Optimización de recursos**: logística, stock, costos.
- **Detección de fraudes y evaluación de riesgos financieros**.

---

## 2. Analítica de Datos

### Tipos de Analítica
- **Descriptiva**: qué ocurrió.
- **Predictiva**: qué podría ocurrir.
- **Prescriptiva**: qué se debería hacer.

### Terminología Clave
`dataset`, `variable`, `modelo`, `algoritmo`, entre otros.

### Categorías de Analítica de Datos
#### a. Analítica Descriptiva – ¿Qué pasó?
- Resume y organiza los datos históricos.
- Uso intensivo de visualizaciones (gráficos, dashboards).

#### b. Analítica Predictiva – ¿Qué puede pasar?
- Utiliza modelos estadísticos y algoritmos de machine learning.
- Métodos principales:
  - **Regresión**: predice valores continuos.
  - **Clasificación**: asigna categorías.
  - **Clustering**: agrupa datos similares.
  - **Reglas de asociación**: detecta elementos que suelen aparecer juntos.
- **Modelos Supervisados**: entrenados con datos etiquetados.  
- **Modelos No Supervisados**: descubren patrones sin etiquetas.

#### c. Analítica Prescriptiva – ¿Qué se debería hacer?
- Propone acciones para influir en resultados futuros.
- Métodos: **Optimización**, **Simulación**, **Análisis de texto**.

---

## 3. Proceso de Ciencia de Datos

### Fases del proceso
1. **Obtención de datos** (bases, APIs, redes sociales).
2. **Curación de datos** (limpieza, tratamiento de valores faltantes).
3. **Ingeniería de características** (feature engineering).
4. **Análisis exploratorio (EDA)**.
5. **Modelado predictivo**.
6. **Evaluación y validación**.
7. **Comunicación de resultados** (visualización, dashboards).

### Roles Comunes
- **Data Scientist** 🧠: desarrolla modelos y visualizaciones.
- **Data Architect** 🏗️: diseña arquitecturas de datos.
- **Data Engineer** ⚙️: crea pipelines de datos.
- **Data Business Analyst** 📊: analiza datos para apoyar decisiones.
- **IT/Infraestructura Engineer** 🖥️: mantiene infraestructura.
- **Usuario (decisor final)** 🧠💼: toma decisiones basadas en datos.

---

## 4. Big Data

### Definición
Conjunto de técnicas para procesar grandes volúmenes de datos provenientes de múltiples fuentes.

### Las 4V del Big Data (Laney, 2001)
- **Volumen**: grandes cantidades de datos.
- **Variedad**: datos en diversos formatos (estructurados, no estructurados).
- **Velocidad**: procesamiento en tiempo real o casi real.
- **Veracidad**: calidad y confiabilidad de los datos.

**Nota**: Un principio clave es GIGO ("Garbage In, Garbage Out"): datos defectuosos producen resultados defectuosos.

---

## 5. Arquitectura de Datos

### Definición
Marco estructural para recolectar, almacenar, transformar, distribuir y utilizar datos.

### Tipos de Arquitectura
- **Tradicional**: data warehouses centralizados.
- **Moderna**: integra data lakes, nube y modelos híbridos.
- **Lambda**: combina procesamiento batch y tiempo real.
- **Kappa**: solo procesamiento en tiempo real.

### Componentes Clave
- **Data Sources**: bases de datos, sensores, redes sociales.
- **Data Storage**: bases relacionales, data lakes, data warehouses.
- **Data Processing**: batch o en tiempo real.
- **Data Integration**: unificación de datos.
- **Data Governance**: calidad, seguridad y trazabilidad.

---

## 6. Business Intelligence (BI)

### Definición
Conjunto de herramientas y prácticas para transformar datos en decisiones de negocio.

### Beneficios
- Mejora la toma de decisiones.
- Optimiza procesos.
- Anticipa cambios de mercado.

### Componentes Clave
- **Recopilación de datos**.
- **Almacenamiento de datos**.
- **Visualización de datos** (gráficos, dashboards).
- **Análisis de datos** (identificación de patrones).

---

## 7. Disponibilidad de Datos

### Definición
Capacidad de acceder y utilizar datos cuando se requieren.

### Requisitos
- **Accesibles**: fácil recuperación.
- **Fiables**: datos precisos y consistentes.
- **Oportunos**: disponibles en el momento justo.

### Importancia
- Permite decisiones informadas.
- Mantiene operaciones fluidas.
- Ayuda al cumplimiento normativo.
- Otorga ventaja competitiva.

### Factores que Afectan la Disponibilidad
- Infraestructura de TI.
- Seguridad de datos.
- Gestión de datos.
- Errores humanos.

### Estrategias de Mejora
- Soluciones de alta disponibilidad.
- Copias de seguridad regulares.
- Planes de recuperación ante desastres.
