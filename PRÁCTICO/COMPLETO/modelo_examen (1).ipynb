#3. Analisis de datos faltantes

```python
# Identificar qué columnas tienen valores nulos.
# Obtener mediante código y expresar qué porcentaje de datos faltan en cada columna.

# Calculate the percentage of missing values in each column
missing_percentage = dfmuestra.isnull().sum() * 100 / len(dfmuestra)

# Print the columns with missing values and their percentages
print("Columns with missing values:")
for column, percentage in missing_percentage.items():
    if percentage > 0:
        print(f"{column}: {percentage:.2f}%")
```

#4. Limpieza y transformacion

```python
# Eliminar columnas con más del 30% de valores nulos.
umbral = 30  # Umbral del 30% de valores nulos

# Filtrar las columnas que cumplen la condición
columnas_a_eliminar = missing_percentage[missing_percentage > umbral].index

# Eliminar las columnas del DataFrame
dfmuestra_filtrado = dfmuestra.drop(columns=columnas_a_eliminar)
```

```python
# Mostrar información del DataFrame filtrado
print("\nDataFrame después de eliminar columnas con más del 30% de valores nulos:")
print('las columnas que cumplen con el requisito para ser eliminadas son:', columnas_a_eliminar)
print(dfmuestra_filtrado.info())
```

```python
# Renombrar las columnas para que no contengan caracteres especiales ni espacios.

# Function to clean column names
def clean_column_name(name):
    # Replace special characters with underscores
    name = re.sub(r'[^\w\s]', '_', name)
    # Replace spaces with underscores
    name = name.replace(' ', '_')
    return name

# Rename columns
new_column_names = {col: clean_column_name(col) for col in dfmuestra_filtrado.columns}
dfmuestra_filtrado = dfmuestra_filtrado.rename(columns=new_column_names)

dfmuestra_filtrado.columns
```

```python
# Convertir las columnas anio, mes, dia en tipo entero (int)

for col in ['anio', 'mes', 'dia']:
    if col in dfmuestra_filtrado.columns:
        try:
            dfmuestra_filtrado[col] = dfmuestra_filtrado[col].astype(int)
        except ValueError as e:
            print(f"Error converting column '{col}' to integer: {e}")
            # Handle the error appropriately, e.g., replace non-numeric values with NaN
            dfmuestra_filtrado[col] = pd.to_numeric(dfmuestra_filtrado[col], errors='coerce')
            #Then convert to integer, ignoring NaN
            dfmuestra_filtrado[col] = dfmuestra_filtrado[col].astype('Int64')
```

```python
dfmuestra_filtrado['facturacion_'] = dfmuestra_filtrado['facturacion_'].str.replace('$', '').str.replace('.', '').str.replace(' ', '')
```

#5. Analisis exploratorio

```python
# Facturación promedio (facturacion_) total.

# Ensure 'facturacion_' column is numeric and handle non-numeric values
dfmuestra_filtrado['facturacion_'] = pd.to_numeric(dfmuestra_filtrado['facturacion_'])

# Calculate the average 'facturacion_'
facturacion_promedio = dfmuestra_filtrado['facturacion_'].mean()

print(f"Facturación promedio: {facturacion_promedio}")

# Calculate the total 'facturacion_'
facturacion_total = dfmuestra_filtrado['facturacion_'].sum()

print(f"Facturación total: {facturacion_total}")
```

```python
# Set display format to avoid scientific notation
pd.options.display.float_format = '{:.2f}'.format

# Group by 'PROVINCIA_ORIGEN_' and calculate the average 'facturacion_'
average_facturacion_by_province = dfmuestra_filtrado.groupby('PROVINCIA_ORIGEN_')['facturacion_'].mean().round(2)

print("Facturación promedio por provincia:")
print(average_facturacion_by_province)

average_facturacion_by_province
```
